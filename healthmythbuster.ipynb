```python
# ---------------------------------------------
# HEALTH MYTH BUSTER 
# ---------------------------------------------

print("=======================================")
print("        HEALTH MYTH BUSTER APP         ")
print("=======================================\n")

# Simple database of myths and facts
false_myths = {
    "vaccines cause autism": "Vaccines DO NOT cause autism. Many scientific studies prove this.",
    "drinking bleach cures diseases": "Bleach is poisonous. Drinking bleach is extremely dangerous.",
    "lemon cures cancer": "No fruit or food can cure cancer. Only medical treatments work.",
    "coronavirus is a hoax": "COVID-19 is real and confirmed by WHO and global health agencies."
}

true_facts = {
    "washing hands prevents infection": "Correct! Handwashing helps prevent viral infections.",
    "exercise improves health": "Yes! Regular exercise helps maintain heart and body health.",
    "drinking water is important": "True. Staying hydrated is essential for the body."
}
# ---------------------------------------------
# FUNCTION TO CHECK THE CLAIM
# ---------------------------------------------
def check_health_claim(claim):
    claim_lower = claim.lower()

    # Check false myths
    for myth in false_myths:
        if myth in claim_lower:
            return "False", false_myths[myth]

    # Check true facts
    for fact in true_facts:
        if fact in claim_lower:
            return "True", true_facts[fact]

    # If not found in both
    return "Uncertain", "No clear information. Please consult reliable medical sources."


# ---------------------------------------------
# MAIN LOOP
# ---------------------------------------------
while True:
    print("\nEnter a health claim (or type 'exit' to quit):")
    user_claim = input(">> ")

    if user_claim.lower() == "exit":
        print("Thank you for using the Health Myth Buster!")
        break

    verdict, explanation = check_health_claim(user_claim)

    print("\nResult:")
    print("Verdict:", verdict)
    print("Explanation:", explanation)
    print("---------------------------------------")
```
